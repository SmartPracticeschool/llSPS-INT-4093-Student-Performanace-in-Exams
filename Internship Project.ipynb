```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dataset.gender = le.fit_transform(dataset.gender)
```

```python
dataset.lunch = le.fit_transform(dataset.lunch )
```

```python
dataset.iloc[:,4] = le.fit_transform(dataset.iloc[:,4])
```

```python
dataset
```

```python
dataset['race/ethnicity'].replace({'group A':'1','group B':'2', 'group C':'3',
                               'group D':'4','group E':'5'},inplace=True)
```

```python
dataset
```

```python
dataset['parental level of education'].replace({'some high school':'1','high school':'1',"associate's degree":'2',
                                        'some college':'3',"bachelor's degree":'4',"master's degree":'5'},inplace=True)
```

```python
dataset
```

```python
passmark = 40
```

```python
dataset['Math_PassStatus'] = np.where(dataset['math score']<passmark, 'F', 'P')
dataset.Math_PassStatus.value_counts()
```

```python
p = sns.countplot(x='parental level of education', data = dataset, hue='Math_PassStatus', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```

```python
dataset['Reading_PassStatus'] = np.where(dataset['reading score']<passmark, 'F', 'P')
dataset.Reading_PassStatus.value_counts()
```

```python
p = sns.countplot(x='parental level of education', data = dataset, hue='Reading_PassStatus', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```

```python
dataset['Writing_PassStatus'] = np.where(dataset['writing score']<passmark, 'F', 'P')
dataset.Writing_PassStatus.value_counts()
```

```python
p = sns.countplot(x='parental level of education', data = dataset, hue='Writing_PassStatus', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```

```python
dataset['OverAll_PassStatus'] = dataset.apply(lambda x : 'F' if x['Math_PassStatus'] == 'F' or 
                                    x['Reading_PassStatus'] == 'F' or x['Writing_PassStatus'] == 'F' else 'P', axis =1)

dataset.OverAll_PassStatus.value_counts()
```

```python
dataset['Total_Marks'] = dataset['math score']+dataset['reading score']+dataset['writing score']
dataset['Percentage'] = dataset['Total_Marks']/3
```

```python
def GetGrade(Percentage, OverAll_PassStatus):
    if ( OverAll_PassStatus == '0'):
        return '0'    
    if ( Percentage >= 80 ):
        return 'A'
    if ( Percentage >= 70):
        return 'B'
    if ( Percentage >= 60):
        return 'C'
    if ( Percentage >= 50):
        return 'D'
    if ( Percentage >= 40):
        return 'E'
    else: 
        return 'F'

dataset['Grade'] = dataset.apply(lambda x : GetGrade(x['Percentage'], x['OverAll_PassStatus']), axis=1)

dataset.Grade.value_counts()
```

```python
dataset
```

```python
dataset['Math_PassStatus'].replace({"P":1,'F':0},inplace=True)
```

```python
dataset['Reading_PassStatus'].replace({"P":1,'F':0},inplace=True)
```

```python
dataset['Writing_PassStatus'].replace({"P":1,'F':0},inplace=True)
```

```python
dataset['OverAll_PassStatus'].replace({"P":1,'F':0},inplace=True)
```

```python
dataset['Grade'].replace({"F":'0','E':'1','D':'2','C':'3','B':'4','A':'5'},inplace=True)
```

```python
dataset
```

```python
x = dataset.iloc[:,:-1].values
x
```

```python
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state =0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10,criterion='mse',random_state=0)
rf.fit(x_train,y_train)
```

```python
from joblib import dump
dump(rf,'Project_deployment.save')
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
